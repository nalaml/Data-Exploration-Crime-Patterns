## 3. Data Cleaning:
 Identify and handle missing data appropriately

```python
# Identify missing data for each column
m_dx = crimex_datasetx.isnull().sum()

# Calculate the percentage of missing data for each column
m_px = (crimex_datasetx.isnull().sum() / len(crimex_datasetx)) * 100

# Create a DataFrame to display the missing data count and percentage
m_ix = pd.DataFrame({
    'Missing Count': m_dx,
    'Missing Percentage': m_px
})

m_ix[m_ix['Missing Count'] > 0]  # Display only columns with missing data
```

```python
# handling the missing data
# to handle the missing data we will drop the the columns with excessive missing values.
clmx_drpx = ['Crm Cd 2', 'Crm Cd 3', 'Crm Cd 4', 'Cross Street']
dx = crimex_datasetx.drop(columns=clmx_drpx,inplace=True )
```

```python
# replacing null values in object column with mode
for column in ['Mocodes','Vict Sex','Vict Descent','Premis Desc','Weapon Desc']:
    if crimex_datasetx[column].isnull().sum() > 0:
        mode_value = crimex_datasetx[column].mode()[0]  # Get the mode value for the column
        crimex_datasetx[column] = crimex_datasetx[column].fillna(mode_value)  # Fill missing values with the mode value
```

```python
# replacing null values in numeric column with mean
for column in ['Premis Cd','Weapon Used Cd','Crm Cd 1']:
    if crimex_datasetx[column].isnull().sum() > 0:
        mean_value = crimex_datasetx[column].mean()  # Get the mean value for the column
        crimex_datasetx[column] = crimex_datasetx[column].fillna(mean_value)  # Fill missing values with the mean value
```

```python
# Identify missing data for each column
m_dx = crimex_datasetx.isnull().sum()

# Calculate the percentage of missing data for each column
m_px = (crimex_datasetx.isnull().sum() / len(crimex_datasetx)) * 100

# Create a DataFrame to display the missing data count and percentage
m_ix = pd.DataFrame({
    'Missing Count': m_dx,
    'Missing Percentage': m_px
})
m_ix
```

```python
# Calculate Q1 (25th percentile) and Q3 (75th percentile) of the 'Vict Age' column
Q1 = cleanx_dtx['Vict Age'].quantile(0.25)
Q3 = cleanx_dtx['Vict Age'].quantile(0.75)
IQR = Q3 - Q1

# condition for outliers
otlrsxcndtnx = ((cleanx_dtx['Vict Age'] < (Q1 - 1.5 * IQR)) |
                      (cleanx_dtx['Vict Age'] > (Q3 + 1.5 * IQR)))

# Select outliers
otlrs = cleanx_dtx.loc[otlrsxcndtnx]

# Remove outliers from the data
crmxdtx_withxoutx_otlrx = cleanx_dtx.loc[~otlrsxcndtnx]

# Calculate the original and new shape of the data
orgxsx = cleanx_dtx.shape
nsx = crmxdtx_withxoutx_otlrx.shape

orgxsx, nsx, otlrs['Vict Age'].describe()
```

```python
from sklearn.preprocessing import OneHotEncoder

# We have used OneHotencoder to convert the categorical values into a one-hot numeric array.

# Initialize OneHotencoder
encdrx = OneHotEncoder(sparse=False)

# Fit and transform the 'AREA NAME' column
encdrx_ax = encdrx.fit_transform(crmxdtx_withxoutx_otlrx[['AREA NAME']])

# The result is a NumPy array. We can convert it to a dataframe and add it back to our main dataframe.
encdrx_ax_dtx = pd.DataFrame(encdrx_ax, columns=encdrx.get_feature_names_out(['AREA NAME']))

# Reset index on our main dataframe to allow concatenation (to avoid index mismatch)
crmxdtx_withxoutx_otlrx.reset_index(drop=True, inplace=True)

# Concatenate the new dataframe with the encoded area names to the original dataframe
crmdtx_encdrx = pd.concat([crmxdtx_withxoutx_otlrx, encdrx_ax_dtx], axis=1)

# Display the first few rows of the dataframe with the encoded categorical data
crmdtx_encdrx.head()
```

Analyze and visualize seasonal patterns in crime data.

```python
import calendar
# We will group the data by year and month again and calculate the number of crimes for each month.
crmx_mntly = crmdtx_encdrx.groupby([crmdtx_encdrx['DATE OCC'].dt.year,
                                                    crmdtx_encdrx['DATE OCC'].dt.month]).size().unstack(level=-1)

# Normalize the data by the number of days in each month
normx_crmx_mntly = crmx_mntly.copy()
for year in normx_crmx_mntly.index:
    for month in normx_crmx_mntly.columns:
        days = calendar.monthrange(year, month)[1]  # Get the number of days in the month
        normx_crmx_mntly.loc[year, month] /= days

# Plot the normalized crime counts
plt.figure(figsize=(14, 7))
for year in normx_crmx_mntly.index:
    plt.plot(normx_crmx_mntly.loc[year], label=f'Year {year}')

plt.title('Seasonal Crime Patterns by Month (Normalized by Days in Month)')
plt.xlabel('Month')
plt.ylabel('Normalized Number of Crimes')
plt.legend()
plt.xticks(ticks=range(1, 13), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.show()
```

Note:
- The plot presented above is designed to illustrate the seasonal patterns of crime, organized by month. To enhance accuracy, the data is normalized by considering the number of days in each month.

- This visualization allows us to scrutinize and discern any recurring seasonal trends within the data. Specifically, we can observe whether there are consistent peaks or troughs in crime rates during certain months across the various years covered by the dataset.

```python
import matplotlib.pyplot as plt
import seaborn as sns
# To analyze the relationship between the day of the week and the frequency of crimes, we first need to extract
# the day of the week from the 'DATE OCC' datetime column. Then we can group by this and the type of crime

# Extract the day of the week (Monday=0, Sunday=6)
crmdtx_encdrx['Day of Week'] = crmdtx_encdrx['DATE OCC'].dt.dayofweek

# Group by the type of crime and the day of the week and count occurrences
crmx_wkx = crmdtx_encdrx.groupby(['Crm Cd Desc', 'Day of Week']).size().unstack(fill_value=0)

# Normalize the counts by the total counts of each crime to get a relative frequency
crmx_wkx_rltvx = crmx_wkx.div(crmx_wkx.sum(axis=1), axis=0)

# plot the relative frequency of a few selected crimes over the days of the week
comx_crmx = [comx_crmx]
comx_crmx.extend(crmx_wkx_rltvx.index[1:3].tolist())

plt.figure(figsize=(14, 7))

for crime in comx_crmx:
    plt.plot(crmx_wkx_rltvx.loc[crime], label=crime)

plt.title('Crime Frequency by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Relative Frequency')
plt.legend()
plt.xticks(ticks=range(0, 7), labels=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
plt.show()
```

## 5. Advanced Analysis (Optional):
Use predictive modeling techniques (e.g., time series forecasting) to predict future
crime trends.

```python
from statsmodels.tsa.arima.model import ARIMA
from pandas.plotting import autocorrelation_plot

# let's create a time series of the total number of crimes by date
crmx_tymsx = crmdtx_encdrx.groupby('DATE OCC').size()

# ARIMA models require stationary data. We can use differencing to stabilize the mean of the time series
crmx_tymsx_dfrncx= crmx_tymsx.diff().dropna()

# Before fitting the model, let's check the autocorrelation plot to choose ARIMA parameters
autocorrelation_plot(crmx_tymsx_dfrncx)
plt.show()

# We will use p=1, d=1, q=1 for ARIMA(1,1,1) as a starting point
# p is the number of lag observations included in the model
# d is the number of times that the raw observations are differenced
# q is the size of the moving average window

# Fit the ARIMA model
armina_mdlx = ARIMA(crmx_tymsx, order=(1, 1, 1))
arima_rsltx = armina_mdlx.fit() # result

# forecast the next 5 periods
frcstx_periods = 5 # forecast period
frcstx = arima_rsltx.get_forecast(steps=frcstx_periods)
frcstx_index = pd.date_range(start=crmx_tymsx.index[-1], periods=frcstx_periods + 1, closed='right')
frcstx_values = frcstx.predicted_mean
frcstx_conf_int = frcstx.conf_int()

# Plot the original data, the fitted values, and the forecast
plt.figure(figsize=(14,7))
plt.plot(crmx_tymsx, label='Original Data')
plt.plot(crmx_tymsx.index, arima_rsltx.fittedvalues, label='Fitted Values')
plt.plot(frcstx_index, frcstx_values, label='frcstx', color='red')
plt.fill_between(frcstx_index, frcstx_conf_int.iloc[:, 0], frcstx_conf_int.iloc[:, 1], color='pink', alpha=0.3)
plt.title('Crime Trends and frcstx')
plt.xlabel('Date')
plt.ylabel('Number of Crimes')
plt.legend()
plt.show()

# Output the forecast values
frcstx_values, frcstx_conf_int

```

```python
# calculate the IQR and identify otlrsx for each area.
crmx_countx_ax = crimex_datasetx.groupby('AREA NAME').size().reset_index(name='Number of Crimes')

# Calculate the IQR for each area and identify otlrsx
def calculate_iqr_otlrsx(data):
    Q1 = data['Number of Crimes'].quantile(0.25)
    Q3 = data['Number of Crimes'].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    otlrsx = data[(data['Number of Crimes'] < lower_bound) | (data['Number of Crimes'] > upper_bound)]
    return otlrsx

otlrsx_by_area = calculate_iqr_otlrsx(crmx_countx_ax)

# any otlrsx
otlrsx_by_area
```

# 10. Predicting Future Trends:
o Employ time series forecasting methods, such as ARIMA or Prophet, to predict
future crime trends based on historical data. Consider incorporating relevant
external factors into your models.


```python
from statsmodels.tsa.arima.model import ARIMA
from pandas.plotting import autocorrelation_plot

# let's create a time series of the total number of crimes by date
crmx_tymsx = crmdtx_encdrx.groupby('DATE OCC').size()

# ARIMA models require stationary data. We can use differencing to stabilize the mean of the time series
crmx_tymsx_dfrncx= crmx_tymsx.diff().dropna()

# Before fitting the model, let's check the autocorrelation plot to choose ARIMA parameters
autocorrelation_plot(crmx_tymsx_dfrncx)
plt.show()

# We will use p=1, d=1, q=1 for ARIMA(1,1,1) as a starting point
# p is the number of lag observations included in the model
# d is the number of times that the raw observations are differenced
# q is the size of the moving average window

# Fit the ARIMA model
armina_mdlx = ARIMA(crmx_tymsx, order=(1, 1, 1))
arima_rsltx = armina_mdlx.fit() # result

# forecast the next 5 periods
frcstx_periods = 5 # forecast period
frcstx = arima_rsltx.get_forecast(steps=frcstx_periods)
frcstx_index = pd.date_range(start=crmx_tymsx.index[-1], periods=frcstx_periods + 1, closed='right')
frcstx_values = frcstx.predicted_mean
frcstx_conf_int = frcstx.conf_int()

# Plot the original data, the fitted values, and the forecast
plt.figure(figsize=(14,7))
plt.plot(crmx_tymsx, label='Original Data')
plt.plot(crmx_tymsx.index, arima_rsltx.fittedvalues, label='Fitted Values')
plt.plot(frcstx_index, frcstx_values, label='frcstx', color='red')
plt.fill_between(frcstx_index, frcstx_conf_int.iloc[:, 0], frcstx_conf_int.iloc[:, 1], color='pink', alpha=0.3)
plt.title('Crime Trends and forecastt')
plt.xlabel('Date')
plt.ylabel('Number of Crimes')
plt.legend()
plt.show()

# Output the forecast values
frcstx_values, frcstx_conf_int
```
